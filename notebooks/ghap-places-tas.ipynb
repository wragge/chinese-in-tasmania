```python
import requests
import pandas as pd
from pathlib import Path
from io import StringIO

s = requests.Session()
```

```python
url = "https://test-ghap.tlcmap.org/search"
url = "https://tlcmap.org/ghap/search"
params = {"state": "TAS", "format": "csv", "searchausgaz": "on", "page": 1, "paging": 100}
```

```python
csv_file = Path("tas-places.csv")
csv_file.unlink(missing_ok=True)
```

```python
dfs = []
more = True
while more:
    #print(params["page"])
    s.get(url, params=params)
    r = s.get(url, params=params)
    df = pd.read_csv(StringIO(r.text))
    if not df.empty:
        dfs.append(df)
        params["page"] += 1
    else:
        more = False
```

```python
df = pd.concat(dfs)
```
