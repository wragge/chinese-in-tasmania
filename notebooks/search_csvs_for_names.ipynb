```python
import pandas as pd
from pathlib import Path
import re
```

```python
df_csvs = pd.read_csv("/home/tim/Workspace/mycode/ozglam-data/src/csvs_for_indexing_checked_20220809.csv")
```

```python
csvs_path = Path("/home/tim/Workspace/mycode/ozglam-data/src/csvs")
```

```python
df_tas = df_csvs.loc[df_csvs["publisher"] == "Libraries Tasmania"]
```

```python
names = Path("nat_names.txt").read_text().split()
```

```python
for csv in df_tas.itertuples():
    csv_dfs = []
    df = pd.read_csv(Path(csvs_path, csv.csv_file), low_memory=False)
    if not pd.isnull(csv.index):
        for col in csv.index.split("|"):
            for name in names:
                df_search = df.dropna(subset=[col]).loc[df.dropna(subset=[col])[col].str.contains(fr"\b{name}\b", regex=True, flags=re.IGNORECASE)]
                if not df_search.empty:
                    csv_dfs.append(df_search)
    try:
        df_all = pd.concat(csv_dfs)
    except ValueError:
        pass
    else:
        df_all.drop_duplicates(inplace=True)
        df_all.to_csv(Path("searches", csv.csv_file.replace(".csv", "search_results.csv")), index=False)
    
```

```python
more_names = []
for csv in df_tas.itertuples():
    df = pd.read_csv(Path(csvs_path, csv.csv_file), low_memory=False)
    if not pd.isnull(csv.index):
        for col in csv.index.split("|"):
            df_search = df.dropna(subset=[col]).loc[df.dropna(subset=[col])[col].str.contains(fr"\bAh\b", regex=True, flags=re.IGNORECASE)]
            more_names += df_search[col].to_list()
```

```python
name_parts = []
for name in more_names:
    name_parts += re.findall(r"\w+", name)
    
name_parts = [n for n in list(set(name_parts)) if len(n) > 1]
```
